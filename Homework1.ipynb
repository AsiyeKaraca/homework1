```python
import time
def geri_sayim(saniye):
    while saniye:
        dakika, saniye_kalan=divmod(saniye,60)
        zaman_format=f"{dakika:02d}:{saniye_kalan:02d}"
        print(zaman_format, end="\r")
        time.sleep(1)
        saniye -= 1
    print("Süre doldu! 00:00")
try:
    kullanici_girisi = int(input("Kaç saniye geri sayalım?"))
    geri_sayim(kullanici_girisi)
except ValueError:
    print("Lütfen geçerli bir sayı giriniz.")
                
```
